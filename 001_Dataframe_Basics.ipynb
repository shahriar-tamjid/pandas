```python
# importing pandas
import pandas as pd

# converting a file into a dataframe
df = pd.read_csv("/content/drive/MyDrive/DataAnalysis/data/weather_data.csv") # give the path

# check the dataframe
df
```

```python
# any python dictionary can also be converted into dataframe
weather_data = {
    'day': ['1/1/2023', '1/2/2023', '1/3/2023', '1/4/2023', '1/5/2023', '1/6/2023'],
    'temperature': [32, 35, 28, 24, 32, 32,],
    'windspeed': [6, 7, 2, 7, 4, 2],
    'event': ['Rain', 'Sunny', 'Snow', 'Snow', 'Rain', 'Sunny']
}

df = pd.DataFrame(weather_data)

# check the dataframe
df
```

```python
# shape means the diemsion of the dataframe
df.shape

# here 6 means the number of rows [number of rows come first]
# here 4 means the number of columns [number of columns come second]
```

```python
# check the number of rows and cols in another way
rows, cols = df.shape

print('rows: ', rows)
print('cols: ', cols)
```

```python
# check the initial 5 rows by using df.head()
df.head()
```

```python
# to check specific number of rows mention the number in the head() method
df.head(2)
```

```python
# check the last 5 rows by using df.tail()
df.tail()
```

```python
# check the very last row
df.tail(1)
```

```python
# check the rows in a specific range
# check rows from 2 to 4
df[2:5]
```

```python
# check the names of all the columns
df.columns
```

```python
# check the contents in a specific column
df['day']  #  df.day
```

```python
# to work with a few cols specify the names in the df
df[['event', 'day', 'temperature']]
```

```python
# get the max temperature from the df
df['temperature'].max()
```

```python
df['temperature'].min()
```

```python
df['temperature'].mean()
```

```python
df['temperature'].std() # standard diviation
```

```python
# to check the quick statistical overview of the df use the describe method
df.describe()
```

```python
# conditionally select rows
df[df['temperature'] == df['temperature'].max()]
```

```python
# to get specific columns after applying conditions
df[['day','temperature']][df['temperature'] == df['temperature'].max()]
```

```python
# setting the a specific column as the main index column of the df
df.set_index('day', inplace = True) # set_index() returns a new df so to replace the changes in the existing df we use the 'inplace = True' property

df
```

```python
# after setting up index the loc method can be used to get data from specific point
df.loc['1/4/2023']
```

```python
# to reset the index into original one use the reset_index method
df.reset_index(inplace = True)

df
```

```python
# it is possible to set a column as index which has duplicate values
# by this the loc method can be used as a key to seach in the entire df
df.set_index('event', inplace=True)
df
```

```python
df.loc['Snow']
```

```python
# converting a list of tuples into a dataframe
weather_data = [
    ('1/1/2023', 32,6,'Rain'),
    ('1/2/2023',35,7,'Sunny'),
    ('1/3/2023',28,2,'Snow')
]

df = pd.DataFrame(weather_data, columns=['day', 'temperature', 'windspeed', 'event'])
df
```

```python
# converting a list of dictionary into a dataframe
weather_data = [
    {'day': '1/1/2023', 'temperature': 32, 'windspeed': 6, 'event': 'Rain'},
    {'day': '1/2/2023', 'temperature': 35, 'windspeed': 7, 'event': 'Sunny'},
    {'day': '1/3/2023', 'temperature': 28, 'windspeed': 2, 'event': 'Snow'}
]

df = pd.DataFrame(weather_data)
df
```
